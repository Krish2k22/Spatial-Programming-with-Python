## Vector Data

### Point

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point
```

```python
df = {
    "city": ["Kolkata\n(22.5726, 88.3639)", "Delhi\n(28.7041, 77.1025)", "Chennai\n(13.0827, 80.2707)"],
    "geometry": [Point(88.3639, 22.5726), Point(77.1025, 28.7041), Point(80.2707, 13.0827)]
}
gdf = gpd.GeoDataFrame(df, crs="EPSG:4326")
print(gdf)
```

```python
fig, ax = plt.subplots()
gdf.plot(ax=ax)
plt.xlim(70, 100)
plt.ylim(10, 30)
plt.grid()

for x, y, city in zip(gdf.geometry.x, gdf.geometry.y, gdf.city):
    ax.annotate(city, xy=(x, y), xytext=(5, -25), textcoords="offset points")
    
plt.show()
```

### Polyline

```python
from shapely.geometry import LineString
```

```python
df = {
    "name": ["Road"],
    "geometry": [LineString([Point(77.1025, 28.7041), Point(88.3639, 22.5726), Point(80.2707, 13.0827)])]
}
gdf = gpd.GeoDataFrame(df, crs="EPSG:4326")
print(gdf)
```

```python
fig, ax = plt.subplots()
gdf.plot(ax=ax)
plt.xlim(70, 100)
plt.ylim(10, 30)
plt.grid()

plt.show()
```

### Polygon

```python
from shapely.geometry import Polygon
```

```python
df = {
    "name": ["Box"],
    "geometry": [Polygon([Point(77.1025, 28.7041), Point(88.3639, 22.5726), Point(80.2707, 13.0827)])]
}
gdf = gpd.GeoDataFrame(df, crs="EPSG:4326")
print(gdf)
```

```python
fig, ax = plt.subplots()
gdf.plot(ax=ax)
plt.xlim(70, 100)
plt.ylim(10, 30)
plt.grid()

plt.show()
```
